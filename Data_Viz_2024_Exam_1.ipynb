#**ข้อ 1**
ความแตกต่างของ ค่าเฉลี่ย ของข้อมูล Chlo_a ทุกๆ station ระหว่างปี 2018 และ ปี 2019

```python
tidy_y2018 = tidy_df[tidy_df['month'].str.contains('2018')]
tidy_y2019 = tidy_df[tidy_df['month'].str.contains('2019')]
```

```python
mean_2018 = tidy_y2018.groupby('station')['Chlo_a'].mean()
mean_2019 = tidy_y2019.groupby('station')['Chlo_a'].mean()
```

```python
diff_mean = mean_2018 - mean_2019
diff_mean
```
